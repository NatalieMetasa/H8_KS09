Sebelum masuk ke classification model, terlebih dahulu cek nilai kolom yang akan digunakan sebagai fitur. Dari hasil visualisasi korelasi sebelumnya diambil 5 kolom yaitu name, distance, surge_multiplier, product_id, dan cab_type.

```python
df['surge_multiplier'].unique()
```

Nilai pada kolom surge_multiplier tidak ada keanehan sama sekali dan bisa diproses.

```python
df['product_id'].unique()
```

Nilai pada kolom product_id memiliki nilai yang sulit untuk dimengerti, oleh sebab itu nilai product_id akan diubah menjadi nilai yang bisa dibaca. Dari hasil visualisasi analisis jenis kendaraan sepertinya ada hubungan dengan product_id.

```python
df['product_id'].value_counts()
```

```python
df_namecab_type
```

Jika dilihat, kedua data ini memang saling berhubungan. Nilai product_id lyft_line memiliki jumlah 51233, dan pada cab_type Lyft name Shared memiliki jumlah yang sama juga. Oleh sebab itu nilai product_id akan diubah menjadi gabungan antara kolom cab_type dan name.

```python
df['product_id'] = (df['cab_type'].astype('str') + " " + df['name'].astype('str')).astype('category')
```

```python
df['product_id'].value_counts()
```

Sekarang nilai pada product_id lebih mudah dipahami.

```python
df['cab_type'].unique()
```

Nilai pada kolom cab_type tidak ada keanehan sama sekali. Setelah cek nilai pada kolom fitur yang akan digunakan, selanjutnya akan dilakukan encoding untuk fitur dengan tipe data categorical.

```python
df['name_cat'] = preprocessing.LabelEncoder().fit_transform(df['name'])
df['product_id_cat'] = preprocessing.LabelEncoder().fit_transform(df['product_id'])
df['cab_type_cat'] = preprocessing.LabelEncoder().fit_transform(df['cab_type'])
```

```python
df[['name', 'name_cat']].value_counts()
```

```python
df[['product_id', 'product_id_cat']].value_counts()
```

```python
df[['cab_type', 'cab_type_cat']].value_counts()
```

Selanjutnya akan dibuat variabel, set train dan test yang akan digunakan nantinya pada saat pembuatan model.

```python
x = df[['distance','surge_multiplier','cab_type_cat','name_cat','product_id_cat']]
y = df['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

# G. Model Definition

**Pemilihan Fitur**

Alasan dipilih kolom name, distance, surge_multiplier, product_id, dan cab_type dikarenakan 5 kolom ini memiliki korelasi yang paling kuat dibandingkan dengan kolom lainnya. Meskipun product_id berasal dari name dan cab_type, tetapi product_id ternyata memberikan pengaruhnya tersendiri. Oleh sebab itu pada saat deployment agar tidak perlu melakukan inputtan berulang, maka yang diinput adalah name, distance, surge_multiplier, dan cab_type, sedangkan nilai product_id akan diambil dari inputtan pada name dan cab_type.

**Model**

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

# H. Model Training

```python
pred_test = model.predict(x_test)
```

```python
test = x_test.copy()
df_test = pd.DataFrame(test)
```

```python
df_test['predicted_price'] = pred_test
```

```python
df_test.head(10)
```

# I. Model Evaluation

```python
print("Accuracy Score: ", str(model.score(x_test, y_test)))
print("Mean Squared Error: ", str(mean_squared_error(y_test, pred_test)))
```

**Kesimpulan**

Model tersebut memiliki akurasi sekitar 50% dengan MSE 42. Artinya model bisa menebak dengan benar harga jasa transportasi sebesar 50% kemungkinan.
