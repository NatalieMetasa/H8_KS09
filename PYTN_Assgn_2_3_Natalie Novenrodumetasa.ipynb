# Confidence Intervals

```python
data_price = data[['BOROUGH','SALE PRICE']]
data_price
```

```python
total = data_price[
    (data['SALE PRICE'] > 1500000) & (data_price['BOROUGH'] == 4)].count()

total = total['BOROUGH']
print('Total properti yang memiliki harga jual lebih besar dari rata-rata di Queens adalah ', total)
```

```python
total_1 = data_price[data_price['SALE PRICE'] > 1500000].count()

total_1 = total_1['SALE PRICE']
print('Jumlah properti yang memiliki harga jual yang melebihi dari rata-rata berjumlah ',total_1)
```

```python
Proporsi_property = total / total_1
Proporsi_property
```

```python
n = total_1
standard_error = np.sqrt(Proporsi_property * (1 - Proporsi_property) / total)
print("standard error:", standard_error)
```

```python
z_score = 1.96
lower = Proporsi_property - z_score* standard_error #lower limit of the CI
upper = Proporsi_property + z_score* standard_error #upper limit of the CI

print("lower CI:",lower)
print("upper CI:",upper)
```

## Hypothesis Testing

Question : Apakah rata-rata harga penjualan di wilayah Queens lebih tinggi aripada wilayah Brooklyn?

H0 : the distributions of all samples are equal

H1 : the distributions of one or more samples are not equa

```python
data_price = data[['BOROUGH','SALE PRICE']]
data_brooklyn = data_price[data_price['BOROUGH'] == 3]
data_brooklyn.head()
```

```python
data_queens = data_price[data_price['BOROUGH'] == 4]
data_queens.head()
```

```python
n1 = len(data_brooklyn)
me1 = data_brooklyn['SALE PRICE'].mean()
std1 = data_brooklyn['SALE PRICE'].std()

print("n1 :", n1)
print("me1 :", me1)
print("std1 :", std1)
```

```python
n1 = len(data_queens)
me1 = data_queens['SALE PRICE'].mean()
std1 = data_queens['SALE PRICE'].std()

print("n1 :", n1)
print("me1 :", me1)
print("std1 :", std1)
```

```python
z_score, pvalue = sm.stats.ztest(data_brooklyn['SALE PRICE'].dropna(),
                                 data_queens['SALE PRICE'].dropna(),alternative='two-sided')
pvalue
```

```python
if pvalue > 0.05:
    print("H0 diterima")
else:
    print("H1 diterima")
```

Conclusion :

Hasil hipotesis diatas menunjukkan bahwa H0 diterima karena hasil p-value lebih dari 0.05
